```python
df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}, inplace=True)

display(df.columns)
```

```python
senioridade_map = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].map(senioridade_map)

display(df['senioridade'].value_counts())
```

```python
contrato_map = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

remoto_map = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

tamanho_empresa_map = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

df['contrato'] = df['contrato'].map(contrato_map)
df['remoto'] = df['remoto'].map(remoto_map)
df['tamanho_empresa'] = df['tamanho_empresa'].map(tamanho_empresa_map)

display(df['contrato'].value_counts())
display(df['remoto'].value_counts())
display(df['tamanho_empresa'].value_counts())
```
